```python
# HTTP 요청 보내는 모듈
import requests
```

```python
url = 'http://www.naver.com'
response = requests.get(url)
```

```python
# 200은 정상
# 400번대는 주소 틀림
# 500번대는 서버측 인증 X
response.status_code
```

```python
# 텍스트 가져오기
response.text
```

```python
from urllib.request import urlopen
import bs4
```

```python
html = urlopen(url)
html
```

```python
bs_obj = bs4.BeautifulSoup(html, 'html.parser')
type(bs_obj)
```

```python
print(bs_obj.prettify())
```
