#### 이미지 검색
- 검색어로 검색된 이미지의 원본이 저장되어 있는 url을 추출
    - 추출된 url로 이미지 다운 저장을 진행 

```python
# 이미지 저장 dir 생성
import os

if not os.path.exists('crawl_data/imgae') :
    os.makedirs('crawl_data/imgae')
```

```python
# 이미지 검색 결과에 맞게 수정 - 이미지 링크만 저장
def get_api_result(cate, key, display=10, start=1, headers=headers) :
    keyword = urllib.parse.quote(key)
    url = create_url(cate, keyword, display, start) # 어떤 cate 검색이든 동일하게 사용
    obj = request_res(url, headers) # 모든 검색에 동일하게 사용
    ### obj 활용 필요 데이터 추출
    ### title, 원 이미지 link, 이미지 크기 
    title = []
    o_link = []
    height=[]
    width=[]
    for item in obj['items'] :
        title.append(item['title'])
        o_link.append(item['link'])
        height.append(item['sizeheight'])
        width.append(item['sizewidth'])
    temp_df = pd.DataFrame({'제목':title,
                            '이미지링크':o_link,
                            '이미지크기1':height,
                            '이미지크기2':width})
    return temp_df
```

```python
def down_image(url, t_name) : 
    # 다운 시간 체크
    start = time.time() # time 객체 생성
    down_path = './crawl_data/imgae/'+t_name
    try : 
        urllib.request.urlretrieve(url,down_path)
    except :
        return '이미지없음'
    print(time.time() - start)
    return down_path
    
```

```python
# 이미지 검색 결과에 맞게 수정 - 저장 가능한 이미지는 저장하고
# 이미지 정보중 저장폴더와 이미지명을 추가
def get_api_result_image(cate, key, display=10, start=1, headers=headers) :
    keyword = urllib.parse.quote(key)
    url = create_url(cate, keyword, display, start) # 어떤 cate 검색이든 동일하게 사용
    obj = request_res(url, headers) # 모든 검색에 동일하게 사용
    ### obj 활용 필요 데이터 추출
    ### title, 원 이미지 link, 이미지 크기 
    title = []
    o_link = []
    height=[]
    width=[]
    i = 1
    for item in obj['items'] :
        link = item['link']
        tmp_name = key+str(i)+".jpg"
        tmp_path = down_image(link,tmp_name) # 이미지 링크를 전달해 다운받고 다운 경로를 반환하는 함수를 활용
        title.append(tmp_path)
        o_link.append(link)
        height.append(item['sizeheight'])
        width.append(item['sizewidth'])
        i=i+1
    temp_df = pd.DataFrame({'저장경로':title,
                            '이미지링크':o_link,
                            '이미지크기1':height,
                            '이미지크기2':width})
    return temp_df
```

```python
# requests 객체를 이용해서 요청
# header 구성이 urllib보다 간단함
import requests
from urllib.parse import urlparse  # 한글 엔코딩
import urllib.parse
import pandas as pd
import urllib.request
import time
```

```python
def create_heaer(c_id, c_s) :
    return {'X-Naver-Client-Id': c_id, 'X-Naver-Client-Secret': c_s}
```

```python
# 요청 url 생성 함수 :json 형식으로 반환
def create_url(cate, keyword, display=10, start=1) :
    b_url = 'https://openapi.naver.com/v1/search/'
    base_url = b_url+cate #'https://openapi.naver.com/v1/search/blog.json'
    param = '?query=' + keyword + '&display=' + str(display) + '&start=' + str(start) # 파라미터 생성 시 공백이 있으면 안됨, 파라미터간의 구분 &
    url = base_url + param
    return url
```

```python
# 요청 값 추출하는 함수 
def request_res(url, headers) :
    result = requests.get(url, headers=headers)
    res_json = result.json()
    return res_json
```

- ex.)250개 이미지를 검색
- 첫번째 호출 : 1~100
- 두번째 호출 : 101~200
- 세번째 호출 : 201~250

- 호출 횟수 결정 : 검색수//100, 단, 나머지가 있으면 호출횟수는 1 증가
- 문서 검색 시작번호 계산 : 현재호출번호 * 100 + 1
- display 수 결정 : 100 : 마지막 호출의 display수는 변경해야 함

```python
# 이미지 검색 결과에 맞게 수정 - 저장 가능한 이미지는 저장하고
# 이미지 정보중 저장폴더와 이미지명을 추가
# 전달되어지는 s_num(추출할 data 수)을 이용해서 서버측 호출을 몇번 진행할 것인지 결정
def get_api_result_image(cate, key, s_num=10, headers=headers) :
    display=100
    keyword = urllib.parse.quote(key)
    ### title, 원 이미지 link, 이미지 크기 
    title = []
    o_link = []
    height=[]
    width=[]
    i = 1

    # 서버 호출 횟수 계산
    page = s_num // 100
    rem = s_num % 100

    # 검색 갯수가 100으로 나눠 떨어지지 않으면 호출횟수(page) 1 증가
    if rem != 0 :
        page=page+1
        
    for p in range(0,page) :
        # 검색 문서 시작 값 결정
        start = p*100+1
        
        # 나머지가 있고 마지막 호출일 경우 나머지만 display로 전달
        if(rem != 0) and (p==page-1) :
            display=rem
            
        url = create_url(cate, keyword, display, start) # 어떤 cate 검색이든 동일하게 사용
        obj = request_res(url, headers) # 모든 검색에 동일하게 사용
        ### obj 활용 필요 데이터 추출
    
        for item in obj['items'] :
            link = item['link']
            tmp_name = key+str(i)+".jpg"
            tmp_path = down_image(link,tmp_name) # 이미지 링크를 전달해 다운받고 다운 경로를 반환하는 함수를 활용
            title.append(tmp_path)
            o_link.append(link)
            height.append(item['sizeheight'])
            width.append(item['sizewidth'])
            i=i+1
    temp_df = pd.DataFrame({'저장경로':title,
                            '이미지링크':o_link,
                            '이미지크기1':height,
                            '이미지크기2':width})
    return temp_df
```
