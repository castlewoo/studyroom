### 모든 섹션 적용 가능하도록 함수로 구성
- 함수 구성시에는 파라미터로 넘겨받아야 하는 값 결정/리턴값 결정
- 6개의 섹션의 헤드라인 뉴스에대한 뉴스 제목, 각 뉴스의 link, 뉴스의 섹션을 df로 저장
- 함수명 : get_headline(섹션별url, 섹션명)
- 반환 : 섹션별 수집 데이터를 dict 구성 후 반환  

```python
df_menu.loc[2]
```

```python
def get_headline(url, section) : 
    # 딕셔너리 구성하기 위해 수집 데이터 저장할 빈 list 생성
    head_title=[]
    head_link=[]

    # 서버 요청 응답(많은 요청이 발생할 수 있으므로 거부 대비 헤더 추가)
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
    res = requests.get(url, headers=headers)

    # bs4객체(파서기) 생성
    html = res.text #requests 객체는 text 속성을 사용해야만 소스가 전달됨
    bs_obj = BeautifulSoup(html,'html.parser')

    # headline 뉴스 추출
    head_L = bs_obj.findAll("div",{"class":"sh_text"})

    for head in head_L :
        head_title.append(head.a.text) #기사제목
        head_link.append(head.a['href']) #기사 세부내용 url
    return({"head_line":head_title, "news_link":head_link,"section":section})

    
```

```python
pd.DataFrame(get_headline(df_menu['link'][1], df_menu['section'][1]))
```

### 모든 섹션에 함수 적용 후 DF로 생성
- get_headline 함수가 dict로 반환하므로
- temp_df 생성 후 누적 df에 결합해서 모두 누적

```python
# 모든 뉴스 저장할 빈 df 생성
# 컬럼명은 함수에서 반환되는 dict의 key와 동일해야 함
head_df = pd.DataFrame({'head_line':[], 'news_link':[],'section':[]})
head_df
```

```python
# 함수는 dict를 반환
for i in range(1,7) :
    temp = pd.DataFrame(get_headline(df_menu['link'][i], df_menu['section'][i])) # 섹션별로 함수가 반환한 data로 구성한 df
    head_df = pd.concat([head_df, temp],axis=0,ignore_index=True)
```
