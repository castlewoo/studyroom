### 네이트 사이트의 메뉴 추출
- 메뉴명과 연결 url을 추출

```python
import pandas as pd
import bs4
import requests
from urllib.request import urlopen
```

```python
url = 'https://www.nate.com'
```

```python
html = urlopen(url)
html
```

```python
html_text = html.read() # 전달된 소스코드만 추출
```

```python
html_text
# bs4 객체
bs_obj = bs4.BeautifulSoup(html_text, 'html.parser')
```

```python
print(bs_obj.prettify())
```

```python
#divGnb
menu = bs_obj.find('div', {'id':'divGnb'})
menu_li = menu.select('li')
menu_li
```

```python
menu_li[0]
menu_li[0].text
menu_li[0].a
menu_li[0].a['href']
```

```python
for li in menu_li :
    print(li.text,li.a['href'])
```

### 네이트 메뉴명과 링크 url df로 생성 후 저장
- 빈 df를 생성 후 df에 data를 행으로 추가
- 각각의 행을 df로 생성 후 concat 함수로 결합

```python
# 빈 df 생성
nate_menu = pd.DataFrame({'메뉴':[], '링크':[]})
nate_menu
```

```python
index = 0
for li in menu_li :
    menu_text = li.text
    menu_link = li.a['href']
    if menu_link == '' or menu_link == 'javascript:;' or menu_text in nate_menu['메뉴'].values :
        continue # 아래 코드 실행하지 않음
    temp = pd.DataFrame({'메뉴':menu_text,'링크':menu_link},index=[index])
    nate_menu = pd.concat([nate_menu, temp])
    index = index +1
```

```python
'뉴스' in nate_menu['메뉴']
'뉴스' in nate_menu['메뉴'].values
```

### 네이버 뉴스 크롤링
- 네이버 뉴스는 네이버 정책에 따라 모든 언론사들의 뉴스가 랜덤하게 배치됨
- 헤드라인 뉴스는 표면적으로는 제공되지 않는다
- https://news.naver.com
- 네이버 메인 페이지는 upgrade 정책에 따라 동적구성되는 경우가 있음
    - 셀레니움을 사용해야 함
- sub 페이지는 정적 구성 이므로 sub인 news로 바로 접근  

```python
# 네이버 뉴스 메인 페이지에서 뉴스섹션명과 링크 url 추출
url = 'https://news.naver.com'
html = urlopen(url)
```

```python
html_text = html.read()
html_text
```

```python
# bs4 객체 생성
bs_obj = bs4.BeautifulSoup(html_text, 'html.parser')
```

```python
print(bs_obj.prettify())
```

```python
bs_obj.select('body > section > header > div.Nlnb._float_lnb > div')[0].ul
type(bs_obj.select('body > section > header > div.Nlnb._float_lnb > div')[0].ul)
```

```python
# 위의 출력 결과인 ul태그를 find함수를 이용해서 추출하시오(개별연습)
ul = bs_obj.find('ul',{'class':"Nlnb_menu_list"})
ul = bs_obj.find('ul')
ul
```

```python
lis = ul.findAll('li')
len(lis)
```

```python
for li in lis : 
    a_tag=li.find('a')
    print(a_tag.text, " : ", a_tag['href'])
```

### 네이버 뉴스 섹션 메뉴의 링크 df로 구성
- 리스트에 각 컬럼 데이터를 모은 후 df로 구성

```python
section = []
link = []
```

```python
for li in lis : 
    a_tag = li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict = {'section':section,'link':link}
naver_news_df = pd.DataFrame(col_dict)
```

```python
naver_news_df.to_csv('./crawl_data/naver_news_section.csv')
```

### 개별연습 
#### 네이버 뉴스 언론사별 section의 **첫번째** 언론사 뉴스에 대해 기사 타이틀과 기사 요약 언론사명 기고날짜를 추출해서 출력하는 코드 작성

```python
# 1. 네이버 뉴스 URL 이용 요청 후 응답 받아오기
url = 'https://news.naver.com'
html = urlopen(url) # url의 서버에게 요청신호를 전송하고 서버가 주는 응답 신호 및 data를 얻어온다
html #HTTPResponse : 요청모듈로 requests 하면 response 신호를 전송하도록 통신규칙이 정해져 있음, HTTP 신호로 요청하고 HTTP 응답을 받아온 객체는 HTTPResponse
```

```python
# 서버가 전달해준 응답 데이터 중 소스코드만 추출
html_text = html.read() # 한번 읽어오면 객체변수안의 소스코드가 삭제됨
# html_text 변수에는 네이버뉴스 첫번째 페이지의 소스코드가 바이너리 형태로 저장됨
```

```python
# 필요 data 추출 하기 위해 파싱객체 생성
# bs4 객체
bs_obj=bs4.BeautifulSoup(html_text,'html.parser')
type(bs_obj) # 소스코드를 저장하고 있는 파싱객체 : bs4의 함수를 사용할 수 있음
```
